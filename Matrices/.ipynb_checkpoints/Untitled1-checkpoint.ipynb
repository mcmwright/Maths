```python
N = 100
z_real, z_imag = meshgrid(linspace(-5, 5, N), linspace(-5, 5, N))
z = z_real + 1j*z_imag

pcolormesh(z_real, z_imag, abs(z + 1/z))
axis('equal')
```

```python
%matplotlib inline

N = 20
z1_real, z1_imag = meshgrid(linspace(-5, 5, 20), linspace(-5, 5, 1000))
z1 = z1_real + 1j*z1_imag

z2_real, z2_imag = meshgrid(linspace(-5, 5, 1000), linspace(-5, 5, 20))
z2 = z2_real + 1j*z2_imag

w1 = z1 + 1/z1
w2 = z2 + 1/z2


plot(real(w1), imag(w1), '-',
    imag(w1), real(w1), '-')
axis('equal')
```

```python
N = 500
z_real, z_imag = meshgrid(linspace(-5, 5, N), linspace(-5, 5, N))
z = z_real + 1j*z_imag

w = z + 1/z

fig = figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(z_real, z_imag, abs(w), facecolors=cm.jet(angle(w)))
ax.set_zlim(0,5)
ax.axis('equal')
```
